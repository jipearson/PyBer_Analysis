```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")


# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Show only January - June
#plt.xlim(0, 5)
#plt.xlim("Jan", "June")

# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot. 
# changing to plt.barh will create a horizontal bar chart.
plt.bar(x_axis, y_axis, color="green", label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot. 
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()

# Create labels for the x and y axes.
plt.ylabel("Date")
plt.xlabel("Fare($)")

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan",label="New York")
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
plt.plot(y_axis, x_axis, 'o',color="red",label="Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Increase the size of the markers by multiplying y-cordinates 
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
```

```python
# Plot the values with marker size coresponding to value 
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 5 for i in y_axis])
```

```python
# Pie Chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, .2 , 0, 0, 0, 0, 0)
plt.subplots(figsize=(10, 10))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', colors=colors)
plt.show()
```
