```python
# Import magic commands
%matplotlib inline

# Import dependencies 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics as stat
from matplotlib.ticker import MultipleLocator

# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
# Create values list for y-axis
avg_fare_list = pyber_ride_df["Avg. Fare ($USD)"]

# Get the standard deviation of the values in the y-axis.
stdev = stat.stdev(avg_fare_list)
stdev
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=3)
plt.show()
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
# Bar Chart
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
#pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar')
plt.show()
```
